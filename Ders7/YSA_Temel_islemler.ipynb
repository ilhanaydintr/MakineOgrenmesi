```python
import numpy as np
girdi_1=np.array([4,5])

agirlik_1={'dugum_1':np.array([2,-1]),
           'dugum_2':np.array([1,1])}

girdi_21=(girdi_1*agirlik_1['dugum_1']).sum()
girdi_22=(girdi_1*agirlik_1['dugum_2']).sum()

cikti_21=1/(1+np.exp(-girdi_21))
cikti_22=1/(1+np.exp(-girdi_22))

girdi_2=np.array([girdi_21,girdi_22])

agirlik_2=np.array([2,1])

girdi_3=(girdi_2*agirlik_2).sum()
cikti_3=1/(1+np.exp(-girdi_3))
cikti_3
```

#Geriye Yayılım Algoritması
image.png

```python
import numpy as np
import matplotlib.pyplot as plt
eps=0.0001
kayip=[]
deger=6
girdi=np.array([5,2,4])
ogrenme_orani=0.001
w=np.array([2,3,1])
hata=(w*girdi).sum()-deger
kayip.append(hata)
while hata>eps:
  egim=girdi*((w*girdi).sum()-deger)
  w=w-ogrenme_orani*egim
  hata=(w*girdi).sum()-deger
  kayip.append(hata)
plt.plot(kayip)
w
```
